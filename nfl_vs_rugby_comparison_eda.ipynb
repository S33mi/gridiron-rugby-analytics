# Gridiron vs Rugby: Stats Comparison EDA

Comparing NFL (American Football) and Rugby Union using recent public data.

Key metrics (approximate averages from 2020–2025 seasons):
- Points per game
- Tackles per game
- Concussions per 1,000 athlete exposures
- Overall injury rate
- Game duration (playing time)
- Player size (height/weight)

```python
#imprt and load/ create data

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns



# Compiled from public sources (William Hill, PubMed, Kaggle NFL, Rugby reports)

data = {

    'Metric': [

        'Points per game (avg)',

        'Tackles per game (avg)',

        'Concussions per 1,000 athlete exposures',

        'Overall injury rate per 1,000 AEs',

        'Game duration (playing time, min)',

        'Avg player height (feet)',

        'Avg player weight (kg)',

        'Players on field'

    ],

    #'NFL': [23.0, 49, 1.0, 4.9, 60, "6'2\"", 244, 11],
    'NFL': [23.0, 49, 1.0, 4.9, 60, 6.1667, 110.68, 11], #244lbs = 110.68kgs, 6'2\""  = 187.86 cm = 6.1667 feet
    #'Rugby Union': [30.0, 120, 2.5, 15.2, 80, "6'2\"", 240, 15]
    'Rugby Union': [30.0, 120, 2.5, 15.2, 80, 6.1667, 108.86, 15] #240 lbs= 108.86 cm

}



df = pd.DataFrame(data)

df
```

```python
#Bar plot comparison

df_melt = df.melt(id_vars='Metric', var_name='Sport', value_name='Value')



plt.figure(figsize=(12, 8))

sns.barplot(x='Metric', y='Value', hue='Sport', data=df_melt)

plt.title('NFL vs Rugby Union – Key Statistics Comparison', fontsize=16)

plt.xticks(rotation=45, ha='right')

plt.ylabel('Value')

plt.tight_layout()

plt.show()
```

## Key Takeaways (2020–2025 averages)

- **Points**: Rugby often higher in open-play leagues (Premiership/Super Rugby ~30 PPG vs NFL ~23 PPG)

- **Tackles**: Rugby much higher (~100–150/game vs NFL ~45–50) due to continuous play

- **Injuries**: Rugby higher overall rate (~15/1000 AEs vs NFL ~5/1000), but concussions debated (rugby ~2.5 vs NFL ~1.0 per 1000 exposures)

- **Player size**: Similar overall; NFL linemen heavier (~315 lbs), rugby forwards ~240–280 lbs

- **Game length**: NFL 60 min play but long clock; rugby 80 min continuous



Sources: Kaggle NFL stats, William Hill comparison, PubMed/AJSM studies, Rugby reports
